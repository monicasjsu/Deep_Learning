As we can see; our target label is an "integer type" which has to be converted into categorical value in order to use in keras.

```
df['quality'] = df['quality'].apply(str)
```

```
df.info()
```

```
output = df["quality"]
input = df.drop(['quality'],axis=1)
output
```

As we are working with multi class classification using keras, we use label encosing to encode all the classes present in the dataset. 

```
encoding = LabelEncoder()
encoding.fit(output)
encode_quality = encoding.transform(output)
df_ylabel = np_utils.to_categorical(encode_quality)
print(df_ylabel)
```

```
def model():
	model = Sequential()
	model.add(Dense(8, input_dim=4, activation='relu'))
	model.add(Dense(3, activation='softmax'))
	model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
	return model
```

```
from keras.wrappers.scikit_learn import KerasClassifier
estimator = KerasClassifier(build_fn=model, epochs=100, batch_size=5, verbose=0)
```

```
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(input,df_ylabel,test_size=0.3,random_state=0)
```

```
def runAlgo(X_train,X_test,y_train,y_test):
    model = Sequential()
    model.add(Dense(10, activation="relu",input_shape=(11,)))
    model.add(Dropout(0.001))
    model.add(Dense(6,activation='softmax'))
    model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
    model.fit(X_train, y_train, epochs=300)
    a = model.evaluate(X_test, y_test)
    return a
```

```
runAlgo(X_train,X_test,y_train,y_test)
```
